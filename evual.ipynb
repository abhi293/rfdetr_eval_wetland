```python
from rfdetr import RFDETRBase
      
model = RFDETRBase()

# Path to the pre-trained COCO weights
model_weights = "rf-detr-base.pth" 

# Corrected Evaluation Call
model.train(
    dataset_dir="dataset",
    output_dir="eval_results",
    
    # FIX: Use 'pretrain_weights' instead of 'resume' to handle the class mismatch
    pretrain_weights=model_weights, 
    
    eval=True,                   # Run evaluation only
    batch_size=8,
    num_workers=2,
    
    # This allows the model to adapt the 91-class COCO head to your 14-class head
    force_no_pretrain=False      
)
```
